# **Cleaning Datasets**

```python
# changing columns titles for telecom dataset 1
columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']
df_tel1.columns = columns
df_tel1.head()
```

```python
# changing columns titles for telecom dataset 2
columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']
df_tel2.columns = columns
df_tel2.head()
```

```python
# changing columns titles for telecom dataset 3
columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']
df_tel3.columns = columns
df_tel3.head()
```

```python
#merge the three datasets using concatenation
tel_data = pd.concat([df_tel1,df_tel2,df_tel3], ignore_index = 1)
tel_data
```

```python
tel_data.columns
  
tel_data.columns = map(str.lower,tel_data.columns)
tel_data.columns

columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']
```

```python
#drop columns with repetitive values
tel_data_new =tel_data.drop(columns = ['country_a' ,	'country_b' ,'cell_on_site'])
tel_data_new.head()
tel_data_new
```

```python
#adding new columns by spilting the date and time columns
tel_data_new[['date', 'time']] = tel_data.date_time.str.split(expand=True)
tel_data_new.head()
```

# **Data Analysis**

```python
# income made by each product
grouped = tel_data_new.groupby(['product'])['value'].sum()
grouped 
```

```python
#usage of a product
grouped1 = tel_data_new.groupby(['product'])['value'].count()
grouped1
```

```python
#Most valuable cell ID in 3 days
Valuable_cell = tel_data_new.groupby('cell_id').count()
Valuable_cell['valueranked'] = Valuable_cell['value'].rank(ascending=1)
Valuable_cell['valueranked']
```
